```python
# analysis libs
import scipy
from scipy import stats, optimize, interpolate
import numpy as np
# plotting libs
import matplotlib as mpl
import matplotlib.pyplot as plt
from scipy.stats import expon, norm
import matplotlib.pyplot as plt


# Example with the exponential distribution
c = 0
lamb = 2

# Create a frozen exponential distribution instance with specified parameters
exp_obj = scipy.stats.expon(c,1/float(lamb))

x_in = np.linspace(0,1,200) # 200 numbers in [0,1], input for ppf()
y_out = exp_obj.pdf(x_in)
plt.plot(x_in,y_out) # graphically check the results of the inverse CDF
```

```python
x_in = np.linspace(0,1,200) # 200 numbers in [0,1], input for ppf()
y_out = exp_obj.ppf(x_in)
plt.plot(y_out, x_in) # graphically check the results of the inverse CDF
```

```python

fig, ax = plt.subplots(1, 1)
x = np.linspace(expon.ppf(0.001),
                expon.ppf(0.999), 1000)
x1 = np.linspace(expon.ppf(0.01),
                expon.ppf(0.99), 100)
ax.plot(x, expon.pdf(x),
       'r-', lw=8, alpha=0.6, label='expon pdf')
ax.plot(x1, expon.pdf(x1),
       'b-', lw=1, alpha=0.6, label='expon pdf')
```
